```python
def task_10(N):
    try:
        N = int(N) 
    except ValueError:
        return 'Введіть ціле число'
    for a in range(1, N):
        for b in range(a, N):
            try:
                if a**2 + b**2 == N:
                    return f'{a}^2 + {b}^2 = {N}'
            except OverflowError:
                return f'Виникла помилка обчислення для a = {a} і b = {b}'
    return f'Число {N} не можна представити квадратами цілих чисел'
```
